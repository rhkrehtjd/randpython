```python
### 6번 ###

def avg(*args):
    total = 0
    count = 0
    for num in args:
        total += num
        count += 1
    return total / count

print(avg(5,3,12,9))
print(avg(2.4,3.2,7.3))
print(avg(10,5))
```

```python
### 7번 ###

import pandas as pd

index = range(2015, 2021)
colnames = ['fisrt', 'second', 'third', 'fourth']
data = [[500,450,520,610],
        [690,700,820,900],
        [1100, 1030,1200,1380],
        [1500, 1650, 1700, 1850],
        [1990,2020,2300,2420],
        [1020,1600,2200,2550]]

df = pd.DataFrame(data, index=index, columns=colnames)
df.to_csv('Sales.csv')
```

```python
### 8번 ###

import matplotlib.pyplot as plt

for i in range(6):
    plt.plot(df.iloc[i,:],label=str(df.index[i]))

plt.title('2015~2020 Quarterly sales')
plt.legend()
plt.xlabel('Quarters')
plt.ylabel('sales')

plt.show()
```
